```python
def load_test_img(file):
    img = image.load_img(file, target_size=(224,224))
    img_array = image.img_to_array(img) # Converting from PIL object into a numpy array

    return img_array.reshape(1, 224,224,3)
```

```python
file = './data/hotdog/test/hot_dog/133015.jpg'
test = load_test_img(file)
```

```python
plt.imshow(image.load_img(file))
```
